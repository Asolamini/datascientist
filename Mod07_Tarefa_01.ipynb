````python
demo_info = pd.DataFrame({'Variável': df_demo.columns, 'Tipo': df_demo.dtypes.values})

demo_info.head(16)
````

````python
qtd_categorias = []

for col in df_demo.columns:
    qtd_categorias.append(df_demo[col].nunique())

demo_info['qtd_categorias'] = qtd_categorias

demo_info.head(16)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df_demo_ok = df_demo.drop(['qtd_filhos','possui_celular','possui_fone_comercial', 'possui_fone', 'possui_email', 'qt_pessoas_residencia', 'mau'], axis=1)

df_demo_ok.head(5)
````

````python
#df_demo_ok.sexo = df_demo_ok.sexo.astype(int)

df_demo_ok.sexo = df_demo_ok.sexo.map({'M': 1, 'F': 0})

df_demo_ok.sexo.value_counts(normalize=True)


````

````python
df_demo_ok.posse_de_veiculo = df_demo_ok.posse_de_veiculo.map({'Y': 1, 'N': 0})
df_demo_ok.posse_de_veiculo.value_counts(normalize=True)
````

````python
df_demo_ok.posse_de_imovel = df_demo_ok.posse_de_imovel.map({'Y': 1, 'N': 0})
df_demo_ok.posse_de_imovel.value_counts(normalize=True)
````

````python
df_demo_ok.head()
````

````python
df_demo_ok.dtypes
````

````python
df_demo_ml = pd.get_dummies(df_demo_ok, columns=['sexo','posse_de_veiculo', 'tipo_renda', 'posse_de_imovel', 'educacao', 'estado_civil', 'tipo_residencia'], dtype=int, drop_first=True)

df_demo_ml.head(10)
````

````python
df_demo_ml.dtypes
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
cross_email = pd.crosstab(df_demo['possui_email'], df_demo['mau']).reset_index()

perc_email = (cross_email.div(cross_email.sum(axis=1), axis=0) * 100).round(2)

perc_email.head(5)
````

````python
cross_veiculo = pd.crosstab(df_demo_ok['posse_de_veiculo'], df_demo['mau']).reset_index()

perc_veiculo = (cross_veiculo.div(cross_veiculo.sum(axis=1), axis=0) * 100).round(2)

perc_veiculo.head(5)
````

A variável mais poderosa é a *posse_veiculo* com 2,21%, já a varíavel *possui_email* apresentou 2,03%, uma diferença de 0,18% p.p.
