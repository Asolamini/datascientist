```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_sinasc = pd.read_csv('b1abb33c4f.csv')

df_sinasc['avg_apgar5'] = df_sinasc['APGAR5'].mean()
```

```python
df_sinasc.head(5)
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=df_sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=df_sinasc)
sns.countplot(ax = axes[1, 0], x='SEXO', data=df_sinasc)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=df_sinasc)
sns.countplot(ax = axes[2, 0], x='ESTCIVMAE', data=df_sinasc)

fig.delaxes(axes[2, 1])

plt.tight_layout()
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='avg_apgar5', data=df_sinasc)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=df_sinasc)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=df_sinasc)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=df_sinasc)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=df_sinasc)

plt.tight_layout()
plt.show()
```
